```python
import konlpy
import nltk
from konlpy.tag import Mecab
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
mecab = Mecab()
```

```python
with open('ko_stopwords.txt', 'rt') as txt_file:
    stop_words = txt_file.readlines()
```

```python
stop_word = []
for idx in range(len(stop_words)):
    stop_word.append(stop_words[idx].replace("\n",""))
```

```python
#2019.01~2020.11.29 IT/Science news content data 
import time

start_time = time.time()
results = []

for idx in range(100):
    example = list(content)[idx]
    #example = [str(i) for i in example]
    word_tokens = mecab.nouns(example)
    word_tokens = [nouns for nouns in word_tokens if nouns not in stop_word]
    results.extend(word_tokens)

print("Total time : ", time.time() - start_time)
print(results)
```

```python
ko = nltk.Text(results , name='분석')
ko = ko.vocab().most_common(30)
dict(ko)
```

```python
result = " ".join(results)
result
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
 
wordcloud = WordCloud(font_path='NanumBarunGothic.ttf', background_color='white').generate(result)
```

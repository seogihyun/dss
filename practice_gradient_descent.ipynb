```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100)
Y = 0.2 * X + 0.5    # w=0.2, b=0.5
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,Y)
plt.show()
```

```python
def plot_prediction(y_pred, y):
    plt.figure(figsize=(8,6))
    plt.scatter(X, y)
    plt.scatter(X, y_pred)
    plt.show()
```

### Gradient Descent 구현

- 업데이트할 W : learning rate * ((Y예측 - Y실제) * X).mean()
- 업데이트할 W : learning rate * ((Y예측 - Y실제) * 1).mean()

```python
W = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate = 0.7

for epoch in range(100):
    y_pred = W * X + b
    
    error = np.abs(y_pred - Y).mean()
    if error == 0.001:
        break
    
    # gradinent descent 계산
    w_grad = learning_rate * ((y_pred - Y)*X).mean()
    b_grad = learning_rate * ((y_pred - Y)).mean()
    
    # W, b값 갱신
    W = W - w_grad
    b = b - b_grad
    
    if epoch & 5 == 0:
        y_pred = W * X + b
        plot_prediction(y_pred, Y)
```

```python
X1 = np.random.rand(100)
X2 = np.random.rand(100)
Y = 0.2 * X1 + 0.1 * X2 + 0.5    # w=0.2,0.1 / b=0.5
```

```python
W1 = np.random.uniform(-1,1)
W2 = np.random.uniform(-1,1)
b = np.random.uniform(-1,1)

learning_rate = 0.7

for epoch in range(100):
    y_pred = (W1 * X1) + (W2 * X2) + b
    
    error = np.abs(y_pred - Y).mean()
    if error == 0.001:
        break
    
    # gradinent descent 계산
    w_grad1 = learning_rate * ((y_pred - Y)*X1).mean()
    w_grad2 = learning_rate * ((y_pred - Y)*X2).mean()
    b_grad = learning_rate * ((y_pred - Y)).mean()
    
    # W, b값 갱신
    W1 = W1 - w_grad1
    W2 = W2 - w_grad2
    b = b - b_grad
    
    if epoch & 5 == 0:
        y_pred = (W1 * X1) + (W2 * X2) + b
        plot_prediction(y_pred, Y)
```
